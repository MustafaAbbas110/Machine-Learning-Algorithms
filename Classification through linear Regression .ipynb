```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array(range(6))
X.shape = (-1,1)
m = len(X)
Y = np.array([0,0,0,1,1,1])
Y.shape = (-1,1)
theta = np.random.random((2,1))
```

```python
X = (X-np.average(X))/np.std(X)
X = np.insert(X,0,1,axis=1)
```

```python
plt.plot(X[:,1],Y,'go')
plt.plot(X[:,1], X.dot(theta), 'b')
plt.grid(True)
plt.show()
```

```python
sqrError = np.square(X.dot(theta))
J = np.sum(sqrError)/(2*m)
```

```python
alpha = 0.3
n = 0
itr = 10
err = 0
J_history = []
J_history.append(J)

while itr > n:
    error = X.dot(theta)-Y
    temp = error.transpose().dot(X)/m
    theta = theta - temp.transpose()
    
    Y_pre = X.dot(theta)
    J_theta = np.sum( np.square(Y_pre-Y) ) / (2*m)
    J_history.append(J_theta)
    
    n+=1
```

```python
plt.plot(X[:,1],Y,'go')
plt.plot(X[:,1], X.dot(theta), 'b')
plt.grid(True)
plt.show()
```

```python
plt.plot(range(len(J_history)), J_history, 'g')
plt.grid(True)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost function')
plt.show()
```
